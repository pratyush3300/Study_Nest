```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_X = "logisticX.csv"
file_Y = "logisticY.csv"
```

```python
data_X = pd.read_csv(file_X, header=None).values
data_Y = pd.read_csv(file_Y, header=None).values.flatten()
```

```python
data_X = (data_X - np.mean(data_X, axis=0)) / np.std(data_X, axis=0)
data_X = np.c_[np.ones(data_X.shape[0]), data_X]
```

```python
def sigmoid_fn(value):
    return 1 / (1 + np.exp(-value))
```

```python
def calculate_cost(parameters, features, target):
    samples = len(target)
    predictions = sigmoid_fn(features @ parameters)
    return (-1/samples) * np.sum(target * np.log(predictions) + (1 - target) * np.log(1 - predictions))
```

```python
def gradient_descent_algo(features, target, parameters, lr, max_iters):
    samples = len(target)
    cost_history = []

    for _ in range(max_iters):
        error = sigmoid_fn(features @ parameters) - target
        parameters -= (lr/samples) * (features.T @ error)
        cost_history.append(calculate_cost(parameters, features, target))

    return parameters, cost_history
```

```python
initial_params = np.zeros(data_X.shape[1])
learning_rate_1 = 0.1
iterations_1 = 1000
trained_params_1, cost_vals_1 = gradient_descent_algo(data_X, data_Y, initial_params, learning_rate_1, iterations_1)
```

```python

def plot_cost_function(cost_values, label):
    plt.plot(range(len(cost_values)), cost_values, label=label)
    plt.xlabel('Iterations')
    plt.ylabel('Cost Value')
    plt.title('Cost Function Convergence')
    plt.legend()
    plt.show()

plot_cost_function(cost_vals_1, f'LR={learning_rate_1}')
```

```python
def visualize_decision_boundary(features, target, parameters):
    plt.figure()
    x_values = np.linspace(min(features[:,1]), max(features[:,1]), 100)
    y_values = -(parameters[0] + parameters[1] * x_values) / parameters[2]
    plt.plot(x_values, y_values, 'g', label='Decision Boundary')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.show()
visualize_decision_boundary(data_X, data_Y, trained_params_1)
```

```python
learning_rate_2 = 5
iterations_2 = 100
temp_params = np.zeros(data_X.shape[1])
trained_params_2, cost_vals_2 = gradient_descent_algo(data_X, data_Y, temp_params, learning_rate_2, iterations_2)
```

```python
plot_cost_function(cost_vals_1[:iterations_2], f'LR={learning_rate_1}')
plot_cost_function(cost_vals_2, f'LR={learning_rate_2}')
```

```python
def compute_metrics(features, target, parameters):
    predictions = sigmoid_fn(features @ parameters) >= 0.5
    true_positive = np.sum((target == 1) & (predictions == 1))
    true_negative = np.sum((target == 0) & (predictions == 0))
    false_positive = np.sum((target == 0) & (predictions == 1))
    false_negative = np.sum((target == 1) & (predictions == 0))

    return np.array([[true_positive, false_positive], [false_negative, true_negative]])

conf_matrix = compute_metrics(data_X, data_Y, trained_params_1)
accuracy = np.trace(conf_matrix) / np.sum(conf_matrix)
precision = conf_matrix[0,0] / (conf_matrix[0,0] + conf_matrix[0,1])
recall = conf_matrix[0,0] / (conf_matrix[0,0] + conf_matrix[1,0])
f1_score = 2 * (precision * recall) / (precision + recall)

print("Confusion Matrix:\n", conf_matrix)
print(f"Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1-score: {f1_score:.4f}")
```
